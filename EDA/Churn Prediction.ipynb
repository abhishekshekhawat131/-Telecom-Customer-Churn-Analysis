#  Three cases:
Normal:          Mean = Median
                 Bell shape, balanced

Right skew:      Mean > Median
                 Long tail on RIGHT
                 Example: price, salary, income

Left skew:       Mean < Median
                 Long tail on LEFT
                 Example: exam scores (most score high, few fail badly)

# Skewness number meaning:
0          = perfectly balanced
0 to 1     = mild skew, acceptable
1 to 2     = moderate skew, watch out
above 2    = high skew, must fix before modeling
-

# Skewness = mean - median / standard deviation (Approximate ans )

# palette
Means color scheme — which colors to use in the plot.
- First bar → steelblue
- Second bar → lightcoral
> Without palette — all bars get the same default color

# hue
Means color by category — split and color bars based on a column's values.
hue=df["gender"]
- Female bar → one color
- Male bar → another color
> Without hue — seaborn does not know which bar gets which color.

# legend
What is Legend?
Legend is a small box on the plot that explains what each color means.
For example:
🟦 Female
🟥 Male
This box is called a legend — it acts as a key/guide for the colors in the plot.

# Why legend=False here?
When you use hue=df["gender"] — seaborn automatically adds a legend box on the plot.
But in a countplot the x axis already shows Female and Male — **so the legend is redundant and unnecessary**.
Having both is like writing the same thing twice on the same chart.
>We are using hue here only because newer seaborn forced us to.

| X axis        | Y axis        | Best Graph                      |
|---------------|---------------|---------------------------------|
| Categorical   | Categorical   | Countplot                       |
| Categorical   | Numerical     | Boxplot / Barplot               |
| Numerical     | Numerical     | Scatter plot / Line plot        |
| Numerical     | alone         | Distribution – Histogram / KDE  |

- Two Categorical columns        → Countplot with hue
- Three or more columns          → Heatmap / Pairplot
- For churn analysis — boxplot always wins because you want to see if churned customers are spread differently, not just compare averages.



# Boxplot vs Barplot: When to use which?

| Feature               | Boxplot                                     | Barplot                                    |
|-----------------------|---------------------------------------------|--------------------------------------------|
| **Shows**             | Spread, median, outliers, full distribution | Only average (or sum / count)              |
| **Use when**          | You want the **full picture** of how data is distributed | You just want to **compare averages** across categories |
| **Best for**          | Numerical vs Categorical                    | Numerical vs Categorical                   |

# Q26:  Create a high_value_customer flag — customers with MonthlyCharges above the 75th percentile. What is their churn rate?
Why this question matters:
-
- Not all customers are equal in business value. A customer paying above the 75th percentile of MonthlyCharges is generating significantly more revenue — losing them hurts the business more.

The hypothesis:
-
High value customers may churn more because:

- They are paying premium prices
- They expect premium service quality
- If service does not match price — they leave

```python
percentile_75 = df['MonthlyCharges'].quantile(0.75)
print(f'75th Percentile: {percentile_75}')
df['High Value Customers'] = df['MonthlyCharges'].apply(lambda x: 1 if x > percentile_75 else 0)

high_value_churn = df.groupby('High Value Customers')['Churn'].value_counts(normalize = True).unstack()*100
print(high_value_churn.round(2))
```

```python
plt.figure(figsize =(7,5))
ax = sns.countplot(x = df['High Value Customers'],hue = df['Churn'],
                   palette = ['steelblue','lightcoral'],legend = True)

for container in ax.containers:
    ax.bar_label(container,fontsize = 11, padding =3)

plt.title("High Value Customer vs Churn")
plt.xlabel("High Value Customer (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
